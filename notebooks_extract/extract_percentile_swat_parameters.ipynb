```python
def get_mannwhitney(df, classes_col, values_col):

    if not len(df[classes_col].unique()) == 2:
        print("Warning, to compare two groups in the statistical tests, we need only two different types in classes")

    group1 = df.loc[df[classes_col] == df[classes_col].unique()[0]][values_col]
    group2 = df.loc[df[classes_col] == df[classes_col].unique()[1]][values_col]

    #perform the Mann-Whitney U test
    result = mannwhitneyu(group1, group2, alternative='two-sided')
    # print(f"The test statistic is {result.statistic}")
    # print(f"The corresponding two-sided p-value is {result.pvalue}")
    
    return { 'statistic': result.statistic, 'pvalue': result.pvalue }
```

```python
def make_heatmap_df(src_df, cross_sel, col_a, col_b, value_col='pvalue'):
    
    hr = np.ones((len(cross_sel), len(cross_sel)))

    for m in cross_sel:
        # print(m)
        num_a = int(m[1])-1

        for idx, row in mw_df_t[[col_a, col_b, value_col]].iterrows():
            if m == row[col_a]:
                num_b = int(row[col_b][1])-1
                hr[num_a,num_b] = row[value_col]
            elif m == row[col_b]:
                num_b = int(row[col_a][1])-1
                hr[num_a,num_b] = row[value_col]
            else:
                pass

    hmap = pd.DataFrame(hr)
    for idx, m in enumerate(cross_sel):
        hmap = hmap.rename(columns={ idx: m})
        hmap = hmap.rename({ idx: m})

    return hmap
```

```python
def plot_heatmap(hmap, param, fu):
    plt.figure(figsize=(6,4))
    plt.ylabel( f"{param} ({fu})", fontsize='x-large')
    plt.title(f"MannWhitney pvalue for {param} ({fu})", fontsize='x-large')
    ax = sns.heatmap(hmap, cmap='viridis', lw=0.5, vmin=0, vmax=0.1)
    plt.show()
```
